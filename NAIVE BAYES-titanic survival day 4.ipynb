# NAIVE BAYES-titanic survival day 4

```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("titanic_survivor.csv")
data.head()
```

```python
data1=data.drop(["PassengerId","Name","SibSp","Parch","Ticket","Cabin","Embarked"],axis=1)
data1.head()
```

```python
data_set=set(data1["Sex"])
data1["Sex"]=data1["Sex"].map({"female":0,"male":1}).astype(int)
```

```python
x=data1.drop(["Survived"],axis=1)
y=data1.Survived
```

```python
x.isna().any()
```

```python
x.Age=x.Age.fillna(x.Age.mean())
```

```python
x.isna().any()
```

```python
x.head(1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=52)
```

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
model.fit(x_train,y_train)
```

```python
x.head()
```

```python
Pclass=int(input("enter persons Pclass no: "))
sex=int(input("mention Gender_class 0 for female and 1 for male: "))
Age=int(input("Enter persons Age: "))
Fare=float(input("Enter persons Fare: "))
person_details=[[Pclass,sex,Age,Fare]]
result=model.predict(person_details)
print(result)
if result==1:
    print("person might be survived")
else:
    print("person might not be survived")
```

```python
y_pred=model.predict(x_test)
print(np.column_stack((y_pred,y_test)))
```

```python
from sklearn.metrics import accuracy_score
print("Acuuracy Score for our model will be {0}% ".format(accuracy_score(y_test,y_pred)*100))
```
