# How to handle missing values

## How to dectect missing values

### df.isna()

* return bool values if null >> False else : True

#### Example :::

```python
df.isna()
```

### df.isnull()

* same like isna function

#### Example :::

```python
df.isnull()
```

## df.isna().sum() & df.isnull().sum()

* count the how many null values in dataframe or series

#### Example :::

```python
df.isna().sum()  #count of null accourding to columns
```

## df.isna().mean() & df.isnull().mean()

#### Example :::

```python
df.isna().mean()*100  # % of null value in columns
```

```python
df.isnull().mean()*100   # % of null value in columns
```

# Fill Missing Values

* we update missing value by -
                            1.mean  >> df['col_name'].mean()
                            2.median >> df['col_name'].median()
                            3.mode   >> df['col_name'].mode()
                            4.max    >> df['col_name'].max()
                            5.min    >>df['col_name'].min()
                            6.statics value  >>100

## df.fillna()

### a) If not specify column name then whole data frame null value updated

```python
df.fillna(100)  #all df null value fill by 100
```

### b).Access columns and then fill null values

```python
df['Cabin'].fillna(100)  #and if updated into orignal data frame use>> inplace=True
```

# Drop rows and columns of if more null values than pass the thresh value

## df.dropna()

### a) If not specify the axis then by default axis is 0 and del all null value present in any rows

#### a)drop rows

```python
df.dropna()   #see index label for undestanding
```

```python
df.dropna(thresh=12)  #thresh means present equal non null values
```

#### Drop columns

* use axis=1 only

```python
df.dropna(axis=1)   #in cabin columns present null values so it is drop
```

```python
df.dropna(thresh=df.shape[0],axis=1)  #if we put thresh is 12 so if every row prsent 12 non null values
```
